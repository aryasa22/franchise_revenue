# Preprocessing

```python
## Drop Unnecessary Columns
df_prep = df_raw.copy()
df_prep = df_prep.drop(columns = ['Unnamed: 0'])

# overview
df_prep.head()
```

## Change Data Type

```python
# month_period
df_month = df_prep.copy()
df_month['month_period'] = pd.to_datetime(df_month['month_period'])

# overview
df_month.sample(10)
```

```python
# all numeric columns
df_num = df_month.copy()
num_cols = ['revenue', 'revenue_avg', 'max_avg', 'revenue_loss', 'loss_perc']

for col in num_cols:
    df_num[col] = df_num[col].apply(lambda x: float(x.replace(',', '')))

# overview
df_num.sample(10)
```

```python
# Recheck fata types
df_num.info()
```

# Analyze

## Rata-rata Revenue Januari Jawa-Bali

```python
df_analyze = df_num.copy()
```

```python
## Mengambil data hanya untuk jawa dan bali
df_jan_jabal = df_analyze[(pd.DatetimeIndex(df_analyze['month_period']).month == 1) & ((df_analyze['Province'].str.contains('Jawa')) 
                                                                                       | (df_analyze['Province'].str.contains('Bali'))
                                                                                      | (df_analyze['Province'].str.contains('DKI'))
                                                                                      | (df_analyze['Province'].str.contains('Yogyakarta'))
                                                                                      | (df_analyze['Province'].str.contains('Banten')))]
```

```python
## Agregasi
df_agg1 = df_jan_jabal.groupby(['Province', 'month_period']).agg({'revenue' : 'mean'})
df_agg1 = df_agg1.reset_index()

df_agg1['year'] = pd.DatetimeIndex(df_agg1['month_period']).year

## mengubah revenue dalam triliun
df_agg1['revenue'] = np.round(df_agg1['revenue']/1000000000000,2)
df_agg1
```

```python
## Visualisasi
fig, ax = plt.subplots(figsize = (15, 10))
sns.barplot(data = df_agg1, x = 'Province', y = 'revenue', hue = 'year',
           palette = ['#B22222', '#0D47A1'], ax = ax)

# legend
plt.legend([], [], frameon = False)

# Background and Frame
fig.patch.set_facecolor('#f7f5f2')
ax.set(frame_on = False)
plt.yticks([])
plt.xticks()
plt.ylabel('')
plt.xlabel('')

# Text and Annotation
plt.text(-0.2, 10, s = '2021',
        color ='#f1f1f1', weight = 600,
        horizontalalignment = 'center',
        fontsize = 9)
plt.text(0.2, 10, s = '2022',
        color ='#f1f1f1', weight = 600,
        horizontalalignment = 'center',
        fontsize = 9)

df_2021 = df_agg1[df_agg1['year'] == 2021]
df_2021 = df_2021.reset_index(drop = True)
df_2022 = df_agg1[df_agg1['year'] == 2022]
df_2022 = df_2022.reset_index(drop = True)

val_2021 = df_2021['revenue']
val_2022 = df_2022['revenue']

for i in range(len(val_2021)):
    plt.text(i-0.2, val_2021[i]-10, s = str(val_2021[i]),
            horizontalalignment = 'center',
            verticalalignment = 'top', weight = 600,
            color = '#f1f1f1', fontsize = 10)
    plt.text(i+0.2, val_2022[i]-10, s = str(val_2022[i]),
            horizontalalignment = 'center',
            verticalalignment = 'top', weight = 600,
            color = '#f1f1f1', fontsize = 10)

# Axes Labels and Ticks
plt.tick_params(bottom = False)
plt.xticks(fontsize = 12, weight = 500)

fig.savefig('revenue_jawa_bali.png', dpi = 400, quality = 100, bbox_inches = 'tight')

plt.show()
```

## Provinsi dengan Pertumbuhan Revenue Tertinggi 2021

```python
# Agregasi
df_prov = df_analyze.copy()
df_prov_agg = df_prov.groupby(['Province', 'month_period']).agg({'revenue' : 'sum'})
df_prov_agg = df_prov_agg.reset_index()
df_prov_agg['pct_change'] = df_prov_agg['revenue'].pct_change() * 100
df_prov_agg = df_prov_agg[pd.DatetimeIndex(df_prov_agg['month_period']).month != 1]
df_prov_avg_change = df_prov_agg.groupby('Province').agg({'pct_change' : 'mean'})
df_prov_avg_change = df_prov_avg_change.reset_index()
df_prov_avg_change = df_prov_avg_change.sort_values('pct_change', ascending = False)
df_prov_avg_change = df_prov_avg_change.reset_index(drop = True)
df_prov_avg_change.head(10)
```

```python
# Mengambil 10 provinsi teratas
df_prov_plot = df_prov_avg_change.iloc[:10]
df_prov_plot['pct_change_billion'] = df_prov_plot['pct_change'] / 1000000000
```

```python
# Visualisasi
fig, ax = plt.subplots(figsize = (10, 7))
sns.barplot(data = df_prov_plot, x = 'pct_change_billion', y = 'Province', orient = 'h',
           ax = ax, color = '#1565C0')

# Background and Frame
fig.patch.set_facecolor('#f7f5f2')
ax.set(frame_on = False)
plt.xticks([])
plt.ylabel('')
plt.xlabel('')

# Axes Labels and Ticks
plt.tick_params(left = False)
plt.yticks(fontsize = 10, weight = 500)

# Text and Annotation
val = df_prov_plot['pct_change_billion']
for i in range(len(val)):
    plt.text(val[i] - 0.1, i, s = str(np.round(val[i], 2)),
            weight = 600, color = '#f1f1f1',
            horizontalalignment = 'right',
            verticalalignment = 'center')

fig.savefig('pertumbuhan.png', dpi = 300, quality = 100, bbox_inches='tight')

plt.show()
```
